```python
batsman = pd.DataFrame()
bowler = pd.DataFrame()
```

```python
batsman = df[df['Avg.x'] > df['Avg.y']]
batsman.head()
```

```python
bowler = df[df['Avg.x'] < df['Avg.y']]
bowler.head()
```

```python
plt.figure(figsize = (25,16))
sns.heatmap(batsman.drop("PLAYER",axis = 1).corr(),cmap = "viridis",annot = True)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

scaler = StandardScaler()
scaler.fit(batsman.drop("PLAYER",axis = 1))
scaler.fit(bowler.drop("PLAYER",axis = 1))

scaled_batsman = scaler.transform(batsman.drop("PLAYER",axis = 1))
scaled_bowler = scaler.transform(bowler.drop("PLAYER",axis = 1))
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_batsman)
pca.fit(scaled_bowler)

batsman_pca = pca.transform(scaled_batsman)
bowler_pca = pca.transform(scaled_bowler)
```
